Import Libraries After **Installing**

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori,association_rules
import warnings
warnings.filterwarnings("ignore")
```

Load Dataset

```python
df = pd.read_csv('CSV Cleaned_Analytics_Dataset.csv')
df
```

Data Preparation 1

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

Data Prepartion 2

```python
df['TotalAmount'] = df['Quantity'] * df['Unit_Price']
```

Data Prepartion 3


```python
today = df['Order_Date'].max() + dt.timedelta(days=1)
```

Calculate RFM Values

```python
rfm = df.groupby('Customer_Id').agg({
    'Order_Date': lambda x: (today - x.max()).days,   # Recency
    'Order_ID': 'nunique',                            # Frequency
    'TotalAmount': 'sum'                               # Monetary
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']

rfm
```

Calculate R Score

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
```

Calculate F Score

```python
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
```

Calculate M Score

```python
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
```

Create RFM Scores (1 to 5 Ranking)

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
```

Combine RFM Scores 1

```python
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + \
                   rfm['F_Score'].astype(str) + \
                   rfm['M_Score'].astype(str)

rfm
```

Combine RFM Scores 2

```python
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

rfm
```

Customer Segmentation

```python
def segment_customer(row):
    if row['R_Score'] == 5 and row['F_Score'] >= 4:
        return 'Champions'
    elif row['F_Score'] >= 4:
        return 'Loyalists'
    elif row['R_Score'] <= 2:
        return 'Hibernating'
    else:
        return 'Regular'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)

rfm
```

Count The  Customer Segment

```python
rfm['Segment'].value_counts()
```

Associaton Rule (Create Basket Format)

```python
basket = df.groupby(['Order_ID','Order_Date','Product_Category'])['Quantity'].sum().unstack().fillna(0)

basket = basket.applymap(lambda x: 1 if x > 0 else 0)

basket
```

Run Apriori

```python
frequent_items = apriori(basket, min_support=0.02, use_colnames=True)

rules = association_rules(frequent_items, metric="lift", min_threshold=1)

rules.sort_values(by='lift', ascending=False)
```

Segment Analysis

```python
segment_analysis = df.groupby('Segment').agg({
    'TotalAmount': 'mean'
})

segment_analysis.sort_values(by='TotalAmount', ascending=False)
```

Calculate LTV monetary  

```python
ltv = rfm.groupby('Segment')['Monetary'].mean().sort_values(ascending=False)

ltv
```

RFM Validation Check

```python
rfm_validation = rfm.groupby('Segment').agg({
    'Monetary': 'mean',
    'Frequency': 'mean'
})

rfm_validation
```

Validate Segments (Statistical Audit)

```python
segment_analysis = rfm.groupby('Segment').agg({
    'Monetary': 'mean',
    'Frequency': 'mean',
    'Recency': 'mean'
})

segment_analysis
```

Caluculate Lifetime Value

```python
rfm['LTV'] = rfm['Monetary'] * rfm['Frequency']

rfm.groupby('Segment')['LTV'].mean()
```
